```python
import math
from collections import defaultdict as d
pos = [0,0]
ip = input().split()
direction = d(int)
for i in range(0,len(ip),2):
    dir = ip[i]
    dis = int(ip[i+1])
    direction[dir] += dis
pos[0] = direction['UP'] - direction['DOWN']
pos[1] = direction['LEFT'] - direction['RIGHT']
res = math.sqrt(pos[0]**2 + pos[1]**2)
print(int(res))
```

```python
def strle(s1,s2):
    len1 = len(s1)
    len2 = len(s2)
    if len1 > len2:
        return s1
    elif len1 < len2:
        return s2
    else:
        return " ".join([s1,s2])

print(strle("Oneee","Three"))
```

```python
class Node:
    def __init__(self,val=None):
        self.val = val
        self.next = None
        
head = Node(10)
head.next = Node(20)
head.next.next = Node(30)
        
ip = list(input().split())
while ip != 'EXIT':
    if ip[0] == 'AB':
        data = ip[1]
        Node(data).next = head.next
        head = Node(data)
    else:
        continue
    
temp = head
while temp is not None:
    print(temp.next)
    temp = temp.next
```
